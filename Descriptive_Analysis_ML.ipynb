```python
# Importing the relevant Python packages:
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import 'mtcars' data from ggplot:

from ggplot import mtcars
```

```python
# Setting the index of mtcars:

mtcars.index = mtcars["name"]
```

```python
# Displaying the header of the data:

mtcars.head()
```

```python
# Calculating the mean of each column:

mean_columns = mtcars.mean()
print(mean_columns)
```

```python
# Calculating the mean of each row (car):

mtcarsMeanCarWise = mtcars.mean(axis=1)
print (mtcarsMeanCarWise)
```

```python
# Calculating the median of each column:

mtcarsMedianCarWise = mtcars.median(axis=0)
print (mtcarsMedianCarWise)
```

```python
# Plotting histograms for each of the columns:

mtcars.hist(bins=10, figsize=(20,15))
plt.show()
```

```python
# Plotting density curves for each column:

for column in mtcars.columns:
    if column != "name":
        mtcars[column].plot(kind="density", figsize= (2,2))
        plt.title(column)
        plt.show()
```

```python
# Plotting the correlation for selected attributes:

from pandas.plotting import scatter_matrix
attributes = ['mpg','disp','hp','wt','qsec']
scatter_matrix(mtcars[attributes], figsize=(12,8))
plt.show()
```

```python
# Calculating the skewness of each column:

skewness = mtcars.skew()
print(skewness)
```

```python
# Calculating the Inter-Quantile spread of 'mpg' attribute:

describe_mpg = mtcars['mpg'].describe()
iqr_mpg = describe_mpg['75%'] - describe_mpg['25%']
print("IQR of mpg:", iqr_mpg)
```

```python
# Plotting the box plot for 'mpg' column:

mtcars.boxplot(column= "mpg")
plt.show()
```

```python
# Calculating the mean, standard deviation, and variance of the 'mpg' column:

mean_mpg = mtcars['mpg'].mean()
std_mpg = mtcars['mpg'].std()
var_mpg = mtcars['mpg'].var()

print(f"Mean of mpg: {mean_mpg}","\n"
      f"Standard Deviation of mpg: {std_mpg}","\n"
      f"Variance of mpg: {var_mpg}")
```

```python
# Extracting cars with automatic and manual transmission:

mtCarsAutomatic = mtcars[mtcars['am'] == 0]

mtCarsManual = mtcars[mtcars['am'] == 1]

print(mtCarsAutomatic)

print(mtCarsManual)
```

```python
# Plotting the box plot for 'mpg' for each transmission type:

mtcars.boxplot(column='mpg', by='am')
plt.show()
```

**Observations from the box plot:

The box plot above shows the distribution of 'mpg' for automatic and manual transmission cars.

We observe that manual transmission cars tend to have higher 'mpg' than automatic transmission cars
i.e.
**the box plot suggests that Manual Cars offer better mileage/ fuel efficency**

What can be observed:
1. Median mpg:
Manual transmission cars (am = 1) generally have a higher median mpg compared to automatic transmission cars (am = 0).

2. Interquartile Range (IQR):
The IQR for manual transmission cars might be smaller or larger compared to automatic transmission cars, indicating the spread of mpg values is different between the two groups.

3. Outliers:
There could be more outliers in one group compared to the other, indicating variability in mpg within that group.

## Generating and Plotting Normally Distributed Data

```python
# Generate 10000 random normally distributed numbers:

norm_data = pd.DataFrame(np.random.normal(size=10000), columns =['Value'])

norm_data.head()
```

```python
# Plotting the density curve along with the median and mean:

# Calculate mean and median
mean_norm = norm_data['Value'].mean()
median_norm = norm_data['Value'].median()

# Plot density curve
ax = norm_data.plot(kind='density',
                    figsize =(10,10),
                    xlim = (-4,4),
                    title='Density Curve of Normally Distributed Data')

# Plot mean and median
plt.axvline(mean_norm,
            color='red',
            linestyle='--',
            linewidth=2,
            label=f'Mean: {mean_norm:.2f}')

plt.axvline(median_norm,
            color='blue',
            linestyle='-',
            linewidth=2,
            label=f'Median: {median_norm:.2f}')

plt.legend()
plt.show()
```

## Generating and Plotting Exponentially Distributed Data

```python
# Generating 10000 exponentially distributed numbers
skewed_data = pd.DataFrame(np.random.exponential(size=10000), columns=['Value'])
```

```python
# Calculate mean and median
mean_exp = skewed_data['Value'].mean()
median_exp = skewed_data['Value'].median()

# Plot density curve
ax = skewed_data.plot(kind='density', figsize=(10, 10), xlim=(-1, 5), title='Density Curve of Exponentially Distributed Data')

# Plot mean and median
plt.axvline(mean_exp, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_exp:.2f}')
plt.axvline(median_exp, color='blue', linestyle='-', linewidth=2, label=f'Median: {median_exp:.2f}')

plt.legend()
plt.show()
```

Observations:

- For normally distributed data, the mean and median are very close to each other.
- For exponentially distributed data, the mean is greater than the median due to the right-skewed nature of the distribution.
